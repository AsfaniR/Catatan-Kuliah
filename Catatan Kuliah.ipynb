```python
#Sorting
#Sorting adalah metode untuk menyortir tipe data list.
#Macam-macam Sorting
#Bubble Sort
#Code:
lst = [5,0,1,10,7,2,5,4,6,9] #data list yang akan disorting
print(lst) #Manampilkan data list yang belum diproses
for k in range(1,len(lst)): #loopingan pertama untuk menentukan banyak sortingan
    for i in range(0,len(lst)-1): #loopingan kedua untuk memproses sorting
        if lst[i]>lst[i+1]: #jika lst ke-i>lst[i+1] maka....
            lst[i],lst[i+1] = lst[i+1],lst[i] #menukar lst[i] dengan lst[i+1]
print(lst) #menampilkan data list yang sudah disorting
#Selection Sort
#Code:
lst = [4,7,1,9,3,10,6] #data list yang akan disorting
print(lst) #menampilkan data sebelum diproses
for i in range(0,len(lst)-1): #loopingan pertama untuk menentukan banyak sortingan dan indeks awal proses
    min=i #menyimpan indeks yang akan diproses
    for j in range(i+1,len(lst)): #loopingan kedua untuk menentukan indeks yang akan dibandingkan dengan indeks awal
        if lst[j]<lst[min]: #jika lst[j]<lst[min]
            min = j #nilai min dirubah menjadi nilai j

    lst[i],lst[min]=lst[min],lst[i] #menukar lst[i] dengan l[min], dan sebaliknya
print(lst) #menampilkan data list setelah disorting
#Insertion Sort
#Code:
lst = [4,7,1,0] #data list yang akan diproses
print(lst) #menampilkan data list sebelum diproses
for i in range(1,len(lst)): #loopingan pertama untuk menentukan 
    nilaiawal = lst[i] #Menyimpan nilai awal
    hole = i #menentukan indeks yang akan dibandingkan
    while lst[hole-1] > nilaiawal and hole > 0   : #ketika lst[hole-1]>nilaiawal, dan hole>0, maka...
        lst[hole] = lst[hole-1] #merubah nilai lst[hole] dengan nilai lst[hole-1]
        hole=hole-1 #untuk memberhentikan loop
    lst[hole] = nilaiawal #merubah nilai lst[hole] dengan nilaiawal
print(lst) #menampilkan data list yang telah diproses
#Quick Sort
#Code:
def QuickSortAwal(A):
    QuickSort(A,0,len(A)-1) #proses untuk memanggil fungsi quicksort

def QuickSort(A,start,end): 
    if start<end: #jika start<end, maka...
        pIndex = Partition(A,start,end) #menentukan nilai pIndex dengan memanggil fungsi Parititon(A,start,end) 
        QuickSort (A,start,pIndex-1) #memanggil fungsi QuickSort (A,start,pIndex-1)
        QuickSort (A,pIndex+1,end) #Memanggil fungsi QuickSort (A,pIndex+1,end)

def Partition(A,start,end):
    pivot = A[end] #menentukan nilai pivot
    pIndex = start #menentukan nilai pIndex
    for i in range(start,end):
        if (A[i] <= pivot): #jika A[i]<=pivot, maka...
            A[i],A[pIndex]=A[pIndex],A[i] #nilai A[i] ditukar dengan nilai A[pIndex]
            pIndex=pIndex+1 #pIndex ditambah 1
    A[pIndex],A[end]=A[end],A[pIndex] #Menukar nilai A[pIndex] dengan nilai A[end]
    return pIndex #mengulang proses (Rekursif)

A=[100,6,500,4,56,2,1,0] #data list sebelum diproses
print(A) #menampilkan data list sebelum diproses
QuickSortAwal(A) #Memproses data dalam fungsi QuickSortAwal
print(A) #menampilkan data list setelah diproses

# Merge Sort
#Code:
def MergedSort(A):
    n = len(A)
    if (n<2):
        return
    mid = len(A)//2
    left = A[:mid]
    right = A[mid:]
    MergedSort(left)
    MergedSort(right)
    i=0
    j=0
    k=0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            A[k]=left[i]
            i=i+1
        else:
            A[k]=right[j]
            j=j+1
        k=k+1

    while i < len(left):
        A[k]=left[i]
        i=i+1
        k=k+1

    while j < len(right):
        A[k]=right[j]
        j=j+1
        k=k+1

Alist= [54,26,93,17,99,77,31,44,55,20,100]
MergedSort(Alist)
print(Alist)

#Shell Sort (penggabungan antara bubble sort dengan insertion sort)
#Code:
def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:

        for startposition in range(sublistcount):
            print(sublistcount)
            gapInsertionSort(alist,startposition,sublistcount)
        sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):
        currentvalue = alist[i]
        position = i
        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap
        alist[position]=currentvalue
        print(alist)       
alist = [35,33,42,10,14,19,27,44]
print(alist)
shellSort(alist)
print(alist)

```
